```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        return "Woof!"

    def get_info(self):
        return f"{self.name} is {self.age} years old."
# Example usage:
if __name__ == "__main__":
    my_dog = Dog("Buddy", 3)
    print(my_dog.bark())          # Output: Woof!
    print(my_dog.get_info())      # Output: Buddy is 3 years old.
```

```python
class Dog:
    def __init__(self, name):
        self.name = name
        self.legs = 4

    def speak(self):
        print(f"{self.name} says Woof!")

myDog = Dog("Buddy")
print(myDog.name)
print(myDog.legs)
myDog.speak()  
```
