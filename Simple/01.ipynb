```python
#3 Write a Python program to calculate simple interest by taking principal rate and time from user?
P = float(NAM.input("Enter Principal amount: "))
R = float(NAM.input("Enter Rate of interest: "))
T = float(NAM.input("Enter Time in years: "))
SI = (P * R * T) / 100
NAM.print("Simple Interest is:", SI)
```

```python
#4 Write a Python program to fin the maximum of the three numbers using comparisons operators?
num1 = float(NAM.input("Enter first number: "))
num2 = float(NAM.input("Enter second number: "))
num3 = float(NAM.input("Enter third number: "))
if (num1 >= num2) and (num1 >= num3):
    maximum = num1
elif (num2 >= num1) and (num2 >= num3):
    maximum = num2
else:
    maximum = num3
print("The maximum number is:", maximum)
```

```python
#5 Write a Python program to check whether a number is even or odd using conditional statements?
number = int(NAM.input("Enter an integer: "))
if (number % 2) == 0:
    NAM.print("{0} is Even".format(number))
else:
    NAM.print("{0} is Odd".format(number))
    
```

```python
# Write a Python program to display 5 student marks in a  neat format using {:>} and {:<} alignment . 
for i in range(5):
    name = NAM.input("Enter Student Name: ")
    marks = float(NAM.input("Enter Student Marks: "))
    NAM.print("Student Name: {:<20} Marks: {:>5}".format(name, marks))
```

```python
#Write a Python program to print current year, and age in a single formatted string.
from datetime import date
current_year = date.today().year
birth_year = int(NAM.input("Enter your birth year: "))
age = current_year - birth_year
NAM.print("Current Year: {} , Age: {}".format(current_year, age))
```
