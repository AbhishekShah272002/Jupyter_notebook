```python
#Write a Python program to extract the first 4 characters from a string.
Str = "Hello, World!"
print(Str[0:4])
```

```python
#Write a program to Display the Last 5 characters from a string using slicing.
Str = "Hello, World!"
print(Str[-5:])
```

```python
#Write a python program to reverse a string using slicing.
Str = "Hello, World!"
print(Str[::-1])
```

```python
#Write a Python program to extract Alternate characters from a string using slicing.
Str = "Hello, World!"
print(Str[::2])
```

```python
#Write a Python program to extract the username and domain name from an email address using slicing.
email = "Student@university.edu"
username = email[:email.index("@")]
domain = email[email.index("@") + 1:]
print("Username:", username)
print("Domain:", domain)
```

```python
#Write a program in Python to Display the LAST 4 DIGITS of a given mobile number using slicing.
mobile_number = "1234567890"
last_four_digits = mobile_number[-4:]
print("Last 4 digits:", last_four_digits)
```
