```python
# We can write and print a String in multiple ways in Python.
# Here are some examples:
First_Name = input("Enter your first name: ")
Last_Name = input("Enter your last name: ")
output = 'Hello, ' + First_Name + ' ' + Last_Name + '!'
print(output)

# Second method using f-strings (Python 3.6+)
print('-----------Second Method-----------')
output = f'Hello, {First_Name} {Last_Name}!'
print(output)

# Third method using format() function
print('-----------Third Method-----------')
output = 'Hello, {} {}!'.format(First_Name, Last_Name)
print(output)
```
