```python
#Using datetime module to print current date and time
from datetime import datetime
now = datetime.now()

print("Current date and time:", now)
```

```python
#Using timeDelta to add days to current date
from datetime import timedelta as td
future_date = now + td  (days=10)
print("Date after 10 days:", future_date)
```

```python
Yesterday = now - td(days=1)
print("Date before 1 day:", Yesterday)
```

```python
from datetime import datetime as dt
print("Today'S date is:", dt.date(now))
```

```python
print("Today'S time is:", dt.time(now))
```

```python
#Print Today'S Day
print("Today'S day is:", now.strftime("%A"))
#Print Today'S Month
print("Today'S month is:", now.strftime("%B"))
#Print Today'S Year
print("Today'S year is:", now.strftime("%Y")) 

```

```python
# Someone Birthday
birthday = input("Enter your birthday date in YYYY-MM-DD format: ")
birth_date = dt.strptime(birthday, "%Y-%m-%d")
print("Your birthday is on a:", birth_date.strftime("%A"))
print("You were born in the month of:", birth_date.strftime("%B"))
print("You were born in the year:", birth_date.strftime("%Y"))
```
