```python
#Using datetime module to print current date and time
from datetime import datetime
now = datetime.now()

print("Current date and time:", now)
```

```python
#Using timeDelta to add days to current date
from datetime import timedelta as td
future_date = now + td  (days=5)
print("Date after 10 days:", future_date)
```

```python
Yesterday = now - td(days=1)
print("Date before 1 day:", Yesterday)
```
