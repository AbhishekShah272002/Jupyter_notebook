```python
'''Write a program to swap two numbers using: 

a temporary variable 
multiple assignments '''
# Swapping using a temporary variable
a = 10
b = 20
temp = a
a = b
b = temp
print("After swapping using a temporary variable:")
print("a =", a)
print("b =", b)
# Swapping using multiple assignments
x = 30
y = 40
x, y = y, x
print("After swapping using multiple assignments:")
print("x =", x)
print("y =", y)
```
