```python
'''Write a program to swap two numbers using: 

a temporary variable 
multiple assignments '''
# Swapping using a temporary variable
a = 10
b = 20
temp = a
a = b
b = temp
print("After swapping using a temporary variable:")
print("a =", a)
print("b =", b)
# Swapping using multiple assignments
x = 30
y = 40
x, y = y, x
print("After swapping using multiple assignments:")
print("x =", x)
print("y =", y)
```

```python
'''Write a program to calculate Simple Interest by taking principal, rate, and time from the user'''
# Taking inputs from the user
principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the rate of interest: "))
time = float(input("Enter the time in years: "))
# Calculating Simple Interest
simple_interest = (principal * rate * time) / 100
print("The Simple Interest is:", simple_interest)
```

```python
'''Write a program to find the maximum of three numbers using comparison operators.  '''
# Taking inputs from the user
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))
# Finding the maximum
if num1 >= num2 and num1 >= num3:
    maximum = num1
elif num2 >= num1 and num2 >= num3:
    maximum = num2
else:
    maximum = num3
print("The maximum of the three numbers is:", maximum)
```

```python
'''Write a program to calculate the area and perimeter of a rectangle. '''
length = float(input("Enter the length of the rectangle: "))
breadth = float(input("Enter the breadth of the rectangle: "))
area = length * breadth
perimeter = 2 * (length + breadth)
print("The area of the rectangle is:", area)
print("The perimeter of the rectangle is:", perimeter)
```

```python
'''Write a program to demonstrate identity and membership operators using lists. '''
list1 = [1, 2, 3, 4, 5]
list2 = [1, 2, 3, 4, 5]
list3 = list1
print("List1:", list1)
print("List2:", list2)
print("List3:", list3)
print("Is List1 identical to List2?", list1 is list2)
print("Is List1 identical to List3?", list1 is list3)
print("Is 3 a member of List1?", 3 in list1)
print("Is 6 a member of List1?", 6 in list1)
```

```python
'''Write a program to print the multiplication table of a given number. '''
number = int(input("Enter a number: "))
for i in range(1, 11):
    print(number, "x", i, "=", number * i)
```

```python
'''Write a program to calculate the factorial of a number using a while loop. '''
number = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= number:
    factorial *= i
    i += 1
print("The factorial of", number, "is", factorial)
```

```python
'''Write a program to generate Fibonacci series up to N terms. '''
n = int(input("Enter the number of terms: "))
a, b = 0, 1
print("Fibonacci series:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
'''Write a program to check whether a number is prime. '''
number = int(input("Enter a number: "))
is_prime = True
if number > 1:
    for i in range(2, number):
        if number % i == 0:
            is_prime = False
            break
if is_prime:
    print(number, "is a prime number.")
else:
    print(number, "is not a prime number.")
```
