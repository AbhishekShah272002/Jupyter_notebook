```python
for n in range(1, 101):
    if n % 15 == 0:
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else: 
        print(n)
```

```python
for number in range(2,100):
    for factor in range(2, int(number**0.5) + 1):
        if number % factor == 0:
            print(f"{number} equals {factor} * {number // factor}")
            break
    else:
        print(f"{number} is a prime number")
```
