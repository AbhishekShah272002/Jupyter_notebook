```python
for number in range(2, 100):
    for fctorial in range(2, int(number**0.5) + 1):
        if number % fctorial == 0:
            break
    else:
        print(f'{number} is a prime number ')
```

```python
def performOperation(num1, num2, opertion):
    if opertion == 'Sum':
        return num1 + num2
    if opertion == 'multiply':
        return num1 * num2

performOperation(5, 3, 'Sum')
```
