```python
# Write a program to reverse a given number take input from user
num = int(input("Enter a number: "))
reverse_num = 0
while num > 0:
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num //= 10
print("Reversed number:", reverse_num)
```

```python
# Write a program to count vowels, consonants, digits, and spaces in a string in short code
text = input("Enter a string: ")
vowels = consonants = digits = spaces = 0
for char in text:
    if char.isalpha():
        if char.lower() in 'aeiou':
            vowels += 1
        else:
            consonants += 1
    elif char.isdigit():
        digits += 1
    elif char.isspace():
        spaces += 1
print(f"Vowels: {vowels}, Consonants: {consonants}, Digits: {digits}, Spaces: {spaces}")
```
