```python
# Write a program to create a list of integers and find the largest and smallest integers in the list.
numbers = [10, 5, 20, 3, 15, 8]
largest = max(numbers)
smallest = min(numbers)
print("Largest number:", largest)
print("Smallest number:", smallest)
```

```python
# Write a program to perform insertion and deletion operations on a list. 
my_list = [1, 2, 3, 4, 5]
# Insertion
my_list.append(6)  # Adding an element at the end of the list
print("List after insertion:", my_list)
# Deletion
my_list.remove(3)  # Removing the element '3' from the list
print("List after deletion:", my_list) 
```

```python
# Write a program to remove duplicates elements from a list and print the unique elements. 
my_list = [1, 2, 3, 4, 5, 2, 3, 6, 7, 8]
unique_list = list(set(my_list))
print("Unique elements:", unique_list)
```
