```python
# Write a program to identify unique visitors to a website from a list of IP addresses.
IP_addresses = [
    "192.168.1.1",
    "192.168.1.2",
    "192.168.1.3",
    "192.168.1.4",
    "192.168.1.5",
    "192.168.1.3",  # Duplicate IP
    "192.168.1.6",
    "192.168.1.7"
]

# Find unique IP addresses
unique_ips = set(IP_addresses)
print("Unique IP addresses:")
for ip in unique_ips: 
    print(ip)
```

```python
# Write a program to compare skills of two job candidates and find 
# common skills
# Unique skills between
candidate1_skills = {"Python", "Java", "SQL", "Machine Learning"}
candidate2_skills = {"Python", "JavaScript", "SQL", "Data Analysis"}

# Find common skills
common_skills = candidate1_skills & candidate2_skills
print("Common skills:", common_skills)

# Find unique skills for each candidate
unique_to_candidate1 = candidate1_skills - candidate2_skills
unique_to_candidate2 = candidate2_skills - candidate1_skills
print("Unique to Candidate 1:", unique_to_candidate1)
print("Unique to Candidate 2:", unique_to_candidate2)
```
