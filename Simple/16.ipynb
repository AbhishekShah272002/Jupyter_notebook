```python
# Write a program to find the student with the highest marks and print his name and marks.
students = {
    "Abhishek": 80,
    "Ram": 90,
    "Shyam": 85
}
highest_marks = 0
top_student = ""
for student, marks in students.items():
    if marks > highest_marks:
        highest_marks = marks
        top_student = student
print(f"The student with the highest marks is {top_student} with {highest_marks} marks")
```
