```python
# Write a function that prints Names, Course, and  University of a student.
def print_student_info(name, course, university):
    print(f"Name: {name}")
    print(f"Course: {course}")
    print(f"University: {university}")
# # Example usage
print_student_info("Alice Smith", "Computer Science", "University of Example")
print_student_info("Bob Johnson", "Mechanical Engineering", "Example State University")

# # Write a function that takes a list of student names and prints each name on a new line.
# def print_student_names(names):
#     for name in names:
#         print(name)
# # # Example usage
# student_names = ["Alice Smith", "Bob Johnson", "Charlie Brown"]
# print_student_names(student_names)
```

```python
# Write a function to print numbers from 1 to 10 and give output.
def print_numbers():
    for i in range(1, 11):
        print(i)
# Example usage
print_numbers()
```

```python
# Write a function add(a,b) that prints the sum of a and b.
def add(a, b):
    result = a + b
    print(f"The sum of {a} and {b} is: {result}")
# Example usage
add(5, 3)
```

```python
# Write a function that find the maximum of two numbers.
def find_maximum(num1, num2):
    if num1 > num2:
        print(f"The maximum of {num1} and {num2} is: {num1}")
    elif num2 > num1:
        print(f"The maximum of {num1} and {num2} is: {num2}")
    else:
        print(f"Both numbers are equal: {num1} and {num2}")
# Example usage
find_maximum(10, 20)
find_maximum(15, 5)
find_maximum(7, 7)
```

```python
# Write a function that checks whether a number is even or odd.
def check_even_odd(number):
    if number % 2 == 0:
        print(f"{number} is an even number.")
    else:
        print(f"{number} is an odd number.")
# Example usage
check_even_odd(10)
check_even_odd(7)
```
