```python
# Write a program to demonstrate all arithmetic operators on two numbers.
a = 10
b = 5
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
print("Floor Division:", a // b)
```

```python
#Write a program using lambda function to find square of a number and print the result.
square = lambda x: x ** 2
number = 4
result = square(number) 
print("square of", number, "is", result)
```

```python
# Write a program to store coordinates of a location (latitude, longitude) and display them in readable format.
latitude = 40.7128
longitude = -74.0060
print(f"Coordinates: {latitude}, {longitude}")
```

```python
# Write a program to maintain attendance of students and count Present and Absent students.
attendance = {
    "Alice": "Present",
    "Bob": "Absent",
    "Charlie": "Present",
    "David": "Absent"
}
present_count = sum(1 for status in attendance.values() if status == "Present")
absent_count = sum(1 for status in attendance.values() if status == "Absent")
print("Present students:", present_count)
print("Absent students:", absent_count)
```

```python
# Write a program to store daily temperatures of a city for 7 days and find Highest, Lowest, and Average temperature
temperatures = [30, 32, 28, 31, 29, 27, 33]
highest_temp = max(temperatures)
lowest_temp = min(temperatures)
average_temp = sum(temperatures) / len(temperatures)
print("Highest Temperature:", highest_temp)
print("Lowest Temperature:", lowest_temp)
print("Average Temperature:", average_temp)
```

```python
# write a program to use break and continue statements while printing numbers from 1 to 20 (skip multiples of 5).
for i in range(1, 21):
    if i % 5 == 0:
        continue
    print(i)
```

```python
# Write a program to use built-in math module to calculate square root and factorial.
import math
number = 16
sqrt_result = math.sqrt(number)
factorial_result = math.factorial(5)
print("Square root of", number, "is", sqrt_result)
print("Factorial of 5 is", factorial_result)
```
